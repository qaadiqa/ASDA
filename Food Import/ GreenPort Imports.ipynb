```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
ghge_df = pd.read_excel('Chinese Food Life Cycle Assessment Database.xlsx', sheet_name = 'Greenhouse Gas Emission')
wu_df = pd.read_excel('Chinese Food Life Cycle Assessment Database.xlsx', sheet_name = 'Water Use')
lu_df = pd.read_excel('Chinese Food Life Cycle Assessment Database.xlsx', sheet_name = 'Land Use')
display(ghge_df.head())
display(wu_df.head())
display(lu_df.head())
```

```python
display(ghge_df['Food group'].unique())
display(wu_df['Food group'].unique())
display(lu_df['Food group'].unique())
```

```python
plt.figure(figsize=(14, 4))
plt.bar(ghge_df['Food group'], ghge_df['GHGE of 1 kg food as consumed\n(kg CO2-eq/kg)'], color='skyblue')
plt.xlabel('Food group')
plt.ylabel('Greenhouse Gas Emissions (GHGE)')
plt.title('GHGE by Food Group')
plt.xticks(rotation=45)


plt.figure(figsize=(14, 4))
plt.bar(wu_df['Food group'], wu_df['WU of 1 kg food as consumed \n(m³/kg)'], color='green')
plt.xlabel('Food group')
plt.ylabel('Water Usage (WU)')
plt.title('WU by Food Group')
plt.xticks(rotation=45)


plt.figure(figsize=(14, 4))
plt.bar(lu_df['Food group'], lu_df['LU of 1 kg food as consumed (㎡/kg)'])
plt.xlabel('Food group')
plt.ylabel('Land Usage (LU)')
plt.title('LU by Food Group')
plt.xticks(rotation=45)
# Show the plot
plt.show()
```

```python
average_ghge_per_group = ghge_df.groupby('Food group')['GHGE of 1 kg food as consumed\n(kg CO2-eq/kg)'].mean().sort_values()
average_wu_per_group = wu_df.groupby('Food group')['WU of 1 kg food as consumed \n(m³/kg)'].mean().sort_values()
average_lu_per_group = lu_df.groupby('Food group')['LU of 1 kg food as consumed (㎡/kg)'].mean().sort_values()
display(average_ghge_per_group)
display(average_wu_per_group)
display(average_lu_per_group)
```

```python
# Perform ANOVA for GHGE between groups
from scipy import stats

groups = [merged_df[merged_df['Food group'] == group]['GHGE of 1 kg food as consumed\n(kg CO2-eq/kg)'] for group in merged_df['Food group'].unique()]
f_stat, p_value = stats.f_oneway(*groups)
print(f'ANOVA result for GHGE: F-statistic = {f_stat}, p-value = {p_value}')
if p_value < 0.5:
    print('Significant differences found in GHCE')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform ANOVA for GHGE (just an example)
model = ols('GHGE_of_1_kg_food_as_consumed ~ C(Food group)', data=merged_df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# If ANOVA is significant, perform Tukey's HSD
if anova_table['PR(>F)'][0] < 0.05:  # Check if p-value is significant
    tukey_result = pairwise_tukeyhsd(merged_df['GHGE_of_1_kg_food_as_consumed'], merged_df['Food group'])
    print(tukey_result.summary())
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import datetime as date
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
titanic_df = pd.read_csv('Titanic.txt')
titanic_df
```

```python
# Drop rows with missing values for simplicity
titanic_df = titanic_df.dropna(subset=['Survived', 'Pclass', 'Age', 'Fare'])

# Define dependent variable (target) and independent variables (predictors)
y = titanic_df['Survived']  # Binary target (0 = No, 1 = Yes)
X = titanic_df[['Pclass', 'Age', 'Sex', 'SibSP', 'Parch']]  # Predictors

# Add a constant for the intercept
X = sm.add_constant(X)
```

```python
# Fit the GLM model
model = sm.GLM(y, X, family=sm.families.Binomial())
result = model.fit()

# Print the summary of the model
print(result.summary())
```

```python
# Predict probabilities
predictions = result.predict(X)

# Convert probabilities to binary outcomes (threshold = 0.5)
predicted_classes = (predictions > 0.5).astype(int)

# Add predictions to the DataFrame
titanic_df['Predicted'] = predicted_classes
print(titanic_df[['Survived', 'Predicted']].head())
```

```python
accuracy = accuracy_score(y, predicted_classes)
print("Accuracy:", accuracy)
```

```python
conf_matrix = confusion_matrix(y, predicted_classes)
print("Confusion Matrix:\n", conf_matrix)
```
